### 1. Is pay inequitable at the company by gender?

### Create a histogram or bar graph from your data.

```python
# Histogram function of Gender (0 for male, 1 for female)
df.hist(column='GenderID', figsize = (16,10),edgecolor='black')
plt.xlabel('Gender')
plt.ylabel('# of Employees')
plt.title('Company Gender Distribution')
```

```python
# Histogram function of Salary
df.hist(column='Salary', figsize = (16,10),edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('# of Employees')
plt.title('Employee Salary Distribution')
```

### Create a boxplot from your data.

```python
# boxplot of Gender (0 for male, 1 for female) with respect to annual Salary (0 for male, 1 for female)
bplot = df.boxplot(by ='GenderID', column =['Salary'], grid = False, figsize = (16,10))
bplot.set_ylabel('Salary')
```

### Create a bivariate plot from your data.

```python
# Convert single column to int dtype.
df['Salary'] = df['Salary'].astype('int')
```

```python
# Scatter plot of salary and gender (0 for male, 1 for female)
df[df['Salary'] < 250000].plot.scatter(x='Salary', y='GenderID', figsize = (16,10))
```

### 5.Create any additional visualizations that will help to answer the question(s) you want to answer.

```python
df_gender=df[['GenderID', 'Salary']]
```

```python
# Plotting the pie chart for above dataframe (0 for male, 1 for female)
df_gender.groupby(['GenderID']).sum().plot(kind='pie', y='Salary')
```

### 6.Summarize your results and make a conclusion. Explain how you arrived at this conclusion and how your visualizations support your conclusion.

##### Results & Findings
$\bullet$ At the company, there are 175 males and 130 females<br>
$\bullet$ Most annual salaries are clustered between 50,000 and 100,000 USD<br>
$\bullet$ Average pay is about the same by gender, however 1 male employee has a salary above 200,000<br>
$\bullet$ The sum of annual salaries is more for males a the company<br>
$\bullet$ Conclusion: The histograms show that most annual salaries are clustered in the 50,000-100,000 USD range for all employees, while the boxplot and scatterplot show this same clustering and separates it out by gender with only 1 male employee in the whole dataset earning above 200,000 USD. The sum of annual salaries is more for males, but this is due more males employed and to the 1 male earning above 200,000 USD (outlier). These dataset visualization findings do not indicate a generlized pay inequality by gender. 

### 2. Which managers have employees with highest & lowest engagement scores?

### Create a histogram or bar graph from your data.

```python
# Histogram function of Engagement Survey score
df.hist(column='EngagementSurvey', figsize = (16,10),edgecolor='black')
plt.xlabel('Engagement Survey Score')
plt.ylabel('# of Employees')
plt.title('Engagement Survey Score Distribution')
```

### Create a boxplot from your data.

```python
# boxplot of Manager with respect to annual Engagement Score
bplot2 = df.boxplot(by ='ManagerID', column =['EngagementSurvey'], grid = False, figsize = (16,10))
bplot2.set_ylabel('Engagement Survey Score')
```

```python
# Calculate the engagement survey median score for the dataset
df['EngagementSurvey'].median()
```

### Create a bivariate plot from your data.

```python
# Scatterplot of Engagement Survey results by ManagerID (note: there are no manager ID's between 31 and 38)
df[df['EngagementSurvey'] < 5].plot.scatter(x='EngagementSurvey', y='ManagerID', figsize = (16,10))
```

### 5.Create any additional visualizations that will help to answer the question(s) you want to answer.

```python
# Hex plot of ManagerID and Engagement Survey Score
df[df['ManagerID'] < 40].plot.hexbin(x='ManagerID', y='EngagementSurvey', gridsize=15, figsize = (16,10))
# Note: the x-axis is ManagerID, but is missing from the chart due to a hex plot bug
```

##### Results & Findings

$\bullet$ In the hexplot, we can see that the engagement scores cluster around 4.3 points and at ManagerID 18 (last x value is 39))<br>
$\bullet$ Conclusion: Most manager's employees have a median employee engagement score of 4.28 as shown in the boxplot, with most engagement scores clustered between 3.5-5.0 in a 1-5 scale. ManagerID 11 Amy Dunn has a longer lower whisker due to multiple low engagement scores, with ManagerID 18 Kelley Spirea having a higher median employee engagement score. From this, we can concluded that lower engagement scores are higher on Amy's team, and it may be worth additional investigation to find out why. Similarly, Kelley's team has higher engagement scores and it might be worth investigate this as well to understand why.
